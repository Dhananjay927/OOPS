##Theory Questions

1- What is Object-Oriented Programming (OOP)?
- Object-Oriented Programming (OOP) is a programming style where you structure your code using objects.Object-oriented programming aims to implement real-world concepts like inheritance, abstraction, polymorphism, encapsulation, etc in programming. The main aim of OOP is to bind together the data and the functions that operate on them so that no other part of the code can access this data except that function

2- What is a class in OOP?
- A class in OOP is a blueprint or template used to create objects. It defines the attributes (data) and methods (behaviors) that the objects will have. Objects created from a class share the same structure but can hold different values.

3- What is an object in OOP?
- An object in OOP is an instance of a class—meaning it is a real, usable thing created from the class blueprint. It contains actual data and can perform actions defined in the class. Each object can have different values for the same attributes.

4- What is the difference between abstraction and encapsulation?
- 1. Abstraction
  - Abstraction means showing only the essential features of an object and hiding unnecessary details.
  - It helps you focus on what a method or object does, not how it does it.
  - It reduces complexity by exposing only a simple interface to the user.
- Encapsulation
  - Encapsulation means binding data and methods together inside a class.
  - It also protects data from unauthorized access using private or protected variables.
  - It helps control how the internal data is accessed or modified (using getters/setters).

5- What are dunder methods in Python?
- Dunder methods (short for Double UNDERscore methods) are special built-in methods in Python whose names start and end with double underscores, like __init__, __str__, __add__, etc.
They allow you to define how your objects should behave with built-in operations (such as printing, adding, comparing, iterating).
These methods are automatically called by Python, not manually.
They make classes more powerful and allow custom objects to behave like built-in types.

6- Explain the concept of inheritance in OOPS.
- Inheritance in OOP is a mechanism that allows one class (child/subclass) to acquire the properties and behaviors (attributes and methods) of another class (parent/superclass).
It helps in reusing code, reducing duplication, and building relationships between classes.

7- What is polymorphism in OOP?
- Polymorphism in OOP means one method name behaving differently depending on the object that calls it.
It allows different classes to have a method with the same name but different implementations.
- It happens in two ways: Method overriding and Method overloading

8- How is encapsulation achieved in Python?
- Encapsulation in Python is achieved by restricting direct access to variables and providing controlled access through methods.
- this is mainly done using:<br>
  1. Private Variables: Prefixing a variable with double underscore makes it private and not directly accessible outside the class.
  2. Getter and Setter Methods: These methods safely read and modify private data.
  3. Using Classes to Bundle Data + Methods: Encapsulation is also achieved simply by placing attributes and methods inside a class.
  4. Protected variables: Prefixing a variable with single underscore makes it protected and only accessible inside the class.

9- What is a constructor in Python?
- A constructor in Python is a special method used to initialize objects when a class is created.
It automatically runs whenever you create an object and is mainly used to set up initial values for attributes.
In Python, the constructor is the method named __init__().

10- What are class and static methods in Python?

modify class state that applies across all instances of the class. Class methods are defined using the
@classmethod decorator.

17- What is the difference between a class variable and an instance variable?

```python
# 1- Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".
class Animal:
  def Speak(self):
    print("Animal Sound")

class Dog(Animal):
  def Speak(self):
    print("Bark!")

obj = Dog()
obj.Speak()
```

```python
# 2- Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius * self.radius

class Rectangle(Shape):
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

obj1 = Circle(4)
print(f"Area of circle is {obj1.area()}")

obj2 = Rectangle(4, 5)
print(f"Area of rectangle is {obj2.area()}")
```

```python
# 3- Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.
class Vehicle:
  def __init__(self, type):
    self.type = type

class Car(Vehicle):
  def __init__(self, type, brand):
    super().__init__(type)
    self.brand = brand

class ElectricCar(Car):
  def __init__(self, type, brand, battery):
    super().__init__(type, brand)
    self.battery = battery

obj = ElectricCar("Car", "Tesla", "50 kWh")
print([obj.type, obj.brand, obj.battery])
```

```python
# 4- Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.
class Bird:
  def fly(self):
    print("Bird is flying")

class Sparrow(Bird):
  def fly(self):
    print("Sparrow is flying")

class Penguin(Bird):
  def fly(self):
    print("Penguin is not flying")

obj1 = Sparrow()
obj1.fly()

obj2 = Penguin()
obj2.fly()
```

```python
# 5- Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount
    return self.__balance

  def withdraw(self, amount):
    if amount > self.__balance:
      return "Insufficient Balance"
    self.__balance -= amount
    return self.__balance

  def check_balance(self):
    return self.__balance

obj = BankAccount(1000)
print(obj.check_balance())
obj.deposit(800)
print(obj.check_balance())
obj.withdraw(1000)
obj.check_balance()
```

```python
# 6- Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().
class Instrument:
  def play(self):
    print("Instrument is playing")

class Guitar(Instrument):
  def play(self):
    print("Guitar is playing")

class Piano(Instrument):
  def play(self):
    print("Piano is playing")

obj = Instrument()
obj.play()

obj = Guitar()
obj.play()

obj = Piano()
obj.play()
```

```python
# 7- Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.
class MathOperations:
  @classmethod
  def add_numbers(cls, a, b):
    return a + b

  @staticmethod
  def subtract_numbers(a, b):
    return a - b

print(MathOperations.add_numbers(2, 3))
print(MathOperations.subtract_numbers(2, 1))
```

```python
# 10- Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y, self.z + other.z)

    def __str__(self):
        return f"({self.x}, {self.y}, {self.z})"

v1 = Vector(1, 2, 3)
v2 = Vector(3, 4, 5)
v3 = v1 + v2
print(v1)
print(v2)
print(v3)
```

```python
# 12- Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.
class Student:
  def __init__(self, name, grades):
    self.name = name
    self.grades = grades

  def average_grade(self):
    return sum(self.grades) / len(self.grades)

obj = Student("Vikram", [10, 20, 30])
print(obj.average_grade())
```

```python
# 13- Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

r = Rectangle()
r.set_dimensions(10, 5)

print("Area of Rectangle:", r.area())
```

```python
# 14- Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
  def __init__(self, hours, rate):
    self.hours = hours
    self.rate = rate

  def calculate_salary(self):
    return self.hours * self.rate

class Manager(Employee):
  def __init__(self, hours, rate, bonus):
    super().__init__(hours, rate)
    self.bonus = bonus

  def calculate_salary(self):
    return super().calculate_salary() + self.bonus

obj1 = Employee(10, 100)
print(obj1.calculate_salary())

obj2 = Manager(10, 100, 1000)
print(obj2.calculate_salary())

```

```python
# 16- Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod
class Animal(ABC):
  @abstractmethod
  def sound(self):
    pass

class Cow(Animal):
  def sound(self):
    print("Cow Sound")

class Sheep(Animal):
  def sound(self):
    print("Sheep Sound")

obj = Cow()
obj.sound()

obj = Sheep()
obj.sound()
```

```python
# 17- Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.
class Book:
  def __init__(self, title, author, year_published):
    self.title = title
    self.author = author
    self.year_published = year_published

  def get_book_info(self):
    return f"Title: {self.title}\nAuthor: {self.author}\nYear Published: {self.year_published}"

obj = Book("The Alchemist", "Paulo Coelho", 1998)
print(obj.get_book_info())
```

```python
# 18- Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
class House:
  def __init__(self, address, price):
    self.address = address
    self.price = price

  def get_house_info(self):
    return f"Address: {self.address}\nPrice: {self.price}"

class Mansion(House):
  def __init__(self, address, price, number_of_rooms):
    super().__init__(address, price)
    self.number_of_rooms = number_of_rooms

  def get_mansion_info(self):
    return f"Address: {self.address}\nPrice: {self.price}\nNumber of Rooms: {self.number_of_rooms}"

obj = House("123 Main St", 1000000)
obj1 = Mansion("123 Main St", 1000000, 5)
print(obj.get_house_info())
print(obj1.get_mansion_info())
```
